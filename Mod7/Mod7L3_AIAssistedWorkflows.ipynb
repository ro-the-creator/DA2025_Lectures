```python
import os
import pickle
import joblib
import numpy as np
import pandas as pd
import streamlit as st

# app.py
import statsmodels.api as sm

# Configuration: adjust if your files are in a different folder
MODEL_DIR = os.environ.get("MODEL_DIR", ".")
PIPE_FILENAME = os.environ.get("PIPE_FILENAME", "pipeline.joblib")
OLS_FILENAME = os.environ.get("OLS_FILENAME", "ols_model.joblib")

pipe_path = os.path.join(MODEL_DIR, PIPE_FILENAME)
ols_path = os.path.join(MODEL_DIR, OLS_FILENAME)

# Load artifacts
pipeline = joblib.load(pipe_path)
with open(ols_path, "rb") as f:
    ols_model = pickle.load(f)

# Infer raw input feature names handled by the pipeline
pre = pipeline.named_steps["preprocessor"]
raw_feature_sets = []
for name, transformer, cols in pre.transformers_:
    # cols can be list of feature names
    if isinstance(cols, (list, tuple)):
        raw_feature_sets.extend(list(cols))
# keep unique & order-preserving
raw_features = []
for c in raw_feature_sets:
    if c not in raw_features:
        raw_features.append(c)

# Defaults (reasonable middle values for UI)
DEFAULTS = {
    "carat": 0.8,
    "depth": 61.75,
    "table": 57.45,
    "x": 5.73,
    "y": 5.73,
    "z": 3.55,
}

st.set_page_config(page_title="Diamond Price Predictor", layout="centered")
st.title("Diamond Price Predictor â€” Streamlit")

st.sidebar.header("Input diamond features")
inputs = {}
for feat in raw_features:
    default = DEFAULTS.get(feat, 0.0)
    if feat == "carat":
        inputs[feat] = st.sidebar.number_input("carat", min_value=0.01, max_value=10.0, value=float(default), step=0.01, format="%.2f")
    else:
        inputs[feat] = st.sidebar.number_input(feat, value=float(default), format="%.3f")

if st.sidebar.button("Predict"):
    # Build single-row raw DataFrame
    raw_df = pd.DataFrame([inputs])[raw_features]

    # Transform with pipeline (ensures transforms fitted only on training data)
    X_proc = pipeline.transform(raw_df)

    # Obtain feature names for transformed matrix
    try:
        feat_names = pre.get_feature_names_out()
    except Exception:
        # fallback: prefix numeric features
        feat_names = np.array([f"num__{c}" for c in raw_features])

    X_df = pd.DataFrame(X_proc, columns=feat_names)
    # Add constant column exactly as model expects
    X_df = sm.add_constant(X_df, has_constant='add')

    # Prediction + confidence interval
    pred = ols_model.get_prediction(X_df)
    pred_df = pred.summary_frame(alpha=0.05)
    predicted_price = float(pred_df["mean"].iloc[0])
    ci_lower = float(pred_df["mean_ci_lower"].iloc[0])
    ci_upper = float(pred_df["mean_ci_upper"].iloc[0])

    st.metric("Predicted Price (USD)", f"${predicted_price:,.0f}")
    st.write(f"95% CI: ${ci_lower:,.0f} â€” ${ci_upper:,.0f}")

    if st.checkbox("Show model coefficients"):
        coef = ols_model.params
        st.write(coef)

    if st.checkbox("Show transformed feature vector"):
        st.write(X_df.T)

st.header("Model card")
with st.expander("Details and limitations", expanded=True):
    st.write("- Model type: statsmodels OLS trained on numeric features transformed by a sklearn ColumnTransformer (StandardScaler).")
    st.write(f"- Features used (raw): {raw_features}")
    st.write(f"- Transformed feature names: {list(feat_names)}")
    try:
        st.write(f"- Trained on n_obs = {int(ols_model.nobs)} rows")
        st.write(f"- R-squared: {ols_model.rsquared:.4f}")
    except Exception:
        pass
    st.write("- Known limits: model is linear, sensitive to outliers and extrapolation. Designed for similar data distribution as original training set. Predictions assume features are within training ranges.")
```
